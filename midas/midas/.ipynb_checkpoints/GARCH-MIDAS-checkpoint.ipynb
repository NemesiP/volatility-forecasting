```python
from volatility import MIDAS, GARCH, T_GARCH
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/peter/Desktop/volatiltiy-forecasting/data/Stocks/AMD.csv')
df['Unnamed: 0'] = df['Unnamed: 0'].astype("datetime64[ms]")
df = df.rename(columns = {"Unnamed: 0": 'Date', 'open': 'Open', 'high': 'High', 'low': 'Low', 'close': 'Close', 'volume': 'Volume'})
df['LogReturn'] = np.log(df.Close).diff().fillna(0)*10
df['Volatility'] = df['LogReturn'] ** 2.0
df = df.iloc[1022:3022, :]
df.set_index(df.Date, inplace = True)
    
vix = pd.read_csv('C:/Users/peter/Desktop/volatiltiy-forecasting/data/Macroeconomic/VIXCLS.csv')
vix = vix[vix.VIXCLS != '.'].reset_index(drop=True)
vix['DATE'] = vix['DATE'].astype("datetime64[ms]")
vix = vix.rename(columns = {'DATE': 'Date', 'VIXCLS': 'VIX'})
vix = vix[(vix.Date >= df.Date.min()) & (vix.Date <= df.Date.max())]
vix['VIX'] = vix['VIX'].astype('float64')
vix.set_index(vix.Date, inplace = True)
    
wti = pd.read_csv('C:/Users/peter/Desktop/volatiltiy-forecasting/data/Macroeconomic/DCOILWTICO.csv')
wti = wti[wti.Value != '.'].reset_index(drop=True)
wti['Date'] = wti['Date'].astype("datetime64[ms]")
wti = wti.rename(columns = {'Date': 'Date', 'Value': 'WTI'})
wti = wti[(wti.Date >= df.Date.min()) & (wti.Date <= df.Date.max())]
wti['WTI'] = wti['WTI'].astype('float64')
wti.set_index(wti.Date, inplace = True)
```

```python
data = pd.concat([df.Volatility, vix.VIX], axis = 1)
data = pd.concat([data, wti.WTI], axis = 1)
data = data.fillna(method = 'bfill')
X = data
y = data.Volatility.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, shuffle = False)
```

## Mixed Data Frequency (MIDAS)

megj.: Itt nem loglikelihood-ot írat ki, hanem a átlagos négyzetes hibát.

```python
selected_lag = 5
restrictions = ['pos', 'pos', 'pos', 'pos', 'pos', 'pos', 'pos', 'pos']

midas = MIDAS(lag = selected_lag)
midas.fit(restrictions, X_train, y_train)
```

```python
y_hat_in = midas.predict(X_train)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data.index[25:1000], y_hat_in[25:], label = 'AMD Volatility')
plt.legend()
plt.title('MIDAS')
plt.tight_layout()
plt.show()
```

# Generalized Autoregressive Conditional Heteroskedasticity (GARCH)
### with Normal Distribution

```python
garch_y = df.LogReturn.values
garch_y[np.where(garch_y == 0.0)] = 1e-5

garch_y_train, garch_y_test = garch_y[:999], garch_y[1000:]
```

```python
restrictions = ['', '01', '01', '01']

garch = GARCH()
garch.fit(restrictions, garch_y_train)
```

```python
garch_y_hat_in = garch.predict(garch_y_train)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data.index[25:1000], garch_y_hat_in[24:], label = 'AMD Volatility')
plt.legend()
plt.title('MIDAS')
plt.tight_layout()
plt.show()
```

# Generalized Autoregressive Conditional Heteroskedasticity (GARCH)
### with Student-t distribution

```python
restrictions = ['', '01', '01', '01', '']

t_garch = T_GARCH()
t_garch.fit(restrictions, garch_y_train)
```

```python
t_garch_y_hat_in = t_garch.predict(garch_y_train)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data.index[25:1000], t_garch_y_hat_in[24:], label = 'AMD Volatility')
plt.legend()
plt.title('MIDAS')
plt.tight_layout()
plt.show()
```
