```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
sim1 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/pgarch_csa/pgcsa_50_sim.csv')
sim2 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/pgarch_csa/pgcsa_100_sim.csv')
```

```python
phi1 = sm.nonparametric.KDEUnivariate(sim1.iloc[:, 1].values).fit()
phi2 = sm.nonparametric.KDEUnivariate(sim2.iloc[:, 1].values).fit()

alp1 = sm.nonparametric.KDEUnivariate(sim1.iloc[:, 2].values).fit()
alp2 = sm.nonparametric.KDEUnivariate(sim2.iloc[:, 2].values).fit()

bet1 = sm.nonparametric.KDEUnivariate(sim1.iloc[:, 3].values).fit()
bet2 = sm.nonparametric.KDEUnivariate(sim2.iloc[:, 3].values).fit()

fig , ax = plt.subplots(3, 1, figsize=(15, 9), tight_layout=True)

ax[0].plot(phi1.support, phi1.density, lw = 3, label = 'N = 50', zorder = 10)
ax[0].plot(phi2.support, phi2.density, lw = 3, label = 'N = 100', marker = 'o', markevery = 10, zorder = 10)
ax[0].set_title(r'$\phi$'+" (Act = 0.9) parameter's density from different samples size")
ax[0].grid(True, zorder = -5)
ax[0].legend(loc = 'best')

ax[1].plot(alp1.support, alp1.density, lw = 3, label = 'N = 50', zorder = 10)
ax[1].plot(alp2.support, alp2.density, lw = 3, label = 'N = 100', marker = 'o', markevery = 10, zorder = 10)
ax[1].set_title(r'$\alpha$'+" (Act = 0.2) parameter's density from different samples size")
ax[1].grid(True, zorder = -5)
ax[1].legend(loc = 'best')

ax[2].plot(bet1.support, bet1.density, lw = 3, label = 'N = 50', zorder = 10)
ax[2].plot(bet2.support, bet2.density, lw = 3, label = 'N = 100', marker = 'o', markevery = 10, zorder = 10)
ax[2].set_title(r'$\beta$'+" (Act = 0.6) parameter's density from different samples size")
ax[2].grid(True, zorder = -5)
ax[2].legend(loc = 'best')

plt.show()
```
