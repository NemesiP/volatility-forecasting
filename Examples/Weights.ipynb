```python
import matplotlib.pyplot as plt
import numpy as np
from weights import beta_, exp_almon_, ewma_
```

```python
fig, ax = plt.subplots(1, 1)

for i in range(1, 11):
    x = np.linspace(0, 1, 100)
    y = beta_().weights([1.0, i], 100)
    ax.plot(x, y, label = 'theta = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Beta Weights')
plt.show()
```

```python
arr = [[3.0, 1.0], [1.5, 2.5], [4.0, 2.0]]

fig, ax = plt.subplots(1, 1)

for i in arr:
    x = np.linspace(0, 1, 100)
    y = beta_().weights(i, 100)
    ax.plot(x, y, label = 'theta1, theta2 = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Beta Weights')
plt.show()
```

```python
arr = [[0.01, -0.001], [0.1, -0.003], [0.008, 0.001]]

fig, ax = plt.subplots(1, 1)

for i in arr:
    x = np.linspace(0, 1, 100)
    y = exp_almon_().weights(i, 100)
    ax.plot(x, y, label = 'theta1, theta2 = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Exponential Almon Weights')
plt.show()
```

```python
arr = np.linspace(0.6, 0.95, 8)

fig, ax = plt.subplots(1,1)

for i in arr:
    x = np.linspace(0, 1, 100)
    y = ewma_().weights(i, 100)
    ax.plot(x, y, label = 'lambda = {}'.format(np.round(i, 2)))
    
ax.legend(loc = 'best')
plt.title('Exponential Weighting')
plt.show()
```
