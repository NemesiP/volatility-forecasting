```python
import sys

sys.path.insert(1, 'C:/Users/peter/Desktop/volatility-forecasting/midas')

import matplotlib.pyplot as plt
import numpy as np
from weights import Beta, ExpAlmon, Exp
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 5))

for i in range(2, 11):
    x = np.linspace(0, 1, 100)
    y = Beta().weights([1.0, i], 100)
    ax.plot(x, y, label = 'theta = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Beta Weights')
plt.show()
```

```python
arr = [[3.0, 1.0], [1.5, 2.5], [4.0, 2.0]]

fig, ax = plt.subplots(1, 1, figsize = (8, 5))

for i in arr:
    x = np.linspace(0, 1, 100)
    y = Beta().weights(i, 100)
    ax.plot(x, y, label = 'theta1, theta2 = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Beta Weights')
plt.show()
```

```python
arr = [[0.01, -0.001], [0.1, -0.003], [0.008, 0.001]]

fig, ax = plt.subplots(1, 1, figsize = (8, 5))

for i in arr:
    x = np.linspace(0, 1, 100)
    y = ExpAlmon().weights(i, 100)
    ax.plot(x, y, label = 'theta1, theta2 = {}'.format(i))

ax.legend(loc = 'best')
plt.title('Exponential Almon Weights')
plt.show()
```

```python
arr = np.linspace(0.6, 0.95, 8)

fig, ax = plt.subplots(1,1, figsize = (8, 5))

for i in arr:
    x = np.linspace(0, 1, 100)
    y = Exp().weights(i, 100)
    ax.plot(x, y, label = 'lambda = {}'.format(np.round(i, 2)))
    
ax.legend(loc = 'best')
plt.title('Exponential Weighting')
plt.show()
```
