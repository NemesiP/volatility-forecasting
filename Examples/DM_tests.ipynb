```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mse_d = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/test_res/mse_values_daily.csv')
mse_d.set_index(mse_d.iloc[:, 0], inplace = True)
mse_d = mse_d.iloc[:, 1:]

mse_td = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/test_res/mse_values_twodaily.csv')
mse_td.set_index(mse_td.iloc[:, 0], inplace = True)
mse_td = mse_td.iloc[:, 1:]

mse_w = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/test_res/mse_values_weekly.csv')
mse_w.set_index(mse_w.iloc[:, 0], inplace = True)
mse_w = mse_w.iloc[:, 1:]

mse_tw = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/test_res/mse_values_twoweekly.csv')
mse_tw.set_index(mse_tw.iloc[:, 0], inplace = True)
mse_tw = mse_tw.iloc[:, 1:]

mse_m = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/test_res/mse_values_monthly.csv')
mse_m.set_index(mse_m.iloc[:, 0], inplace = True)
mse_m = mse_m.iloc[:, 1:]
```

```python
mse_mean = pd.DataFrame(columns = mse_d.columns)
mse_mean.loc['Daily'] = np.nanmean(mse_d, axis = 0)
mse_mean.loc['Two_Daily'] = np.nanmean(mse_td, axis = 0)
mse_mean.loc['Weekly'] = np.nanmean(mse_w, axis = 0)
mse_mean.loc['Two_weekly'] = np.nanmean(mse_tw, axis = 0)
mse_mean.loc['Monthly'] = np.nanmean(mse_m, axis = 0)
```

```python
mse_mean
```

```python
plt.figure(figsize = (15,5))
sns.kdeplot(mse_d['GARCH_MIDAS_norm_vs_EWMA'], label = 'Daily')
sns.kdeplot(mse_td['GARCH_MIDAS_norm_vs_EWMA'], label = 'Two_Daily')
sns.kdeplot(mse_w['GARCH_MIDAS_norm_vs_EWMA'], label = 'Weekly')
sns.kdeplot(mse_tw['GARCH_MIDAS_norm_vs_EWMA'], label = 'Two_Weekly')
sns.kdeplot(mse_m['GARCH_MIDAS_norm_vs_EWMA'], label = 'Monthly')
plt.xlim(-5, 5)
plt.legend(loc = 'best')
plt.show()
```
