```python
mse_pos_pct.loc['daily'] = mse_d[(mse_d >= 1.96)].count()/mse_d.shape[0] * 100
mse_neg_pct.loc['daily'] = mse_d[(mse_d <= -1.96)].count()/mse_d.shape[0] * 100
mse_pos_pct.loc['two_daily'] = mse_td[(mse_td >= 1.96)].count()/mse_td.shape[0] * 100
mse_neg_pct.loc['two_daily'] = mse_td[(mse_td <= -1.96)].count()/mse_td.shape[0] * 100
mse_pos_pct.loc['weekly'] = mse_w[(mse_w >= 1.96)].count()/mse_w.shape[0] * 100
mse_neg_pct.loc['weekly'] = mse_w[(mse_w <= -1.96)].count()/mse_w.shape[0] * 100
mse_pos_pct.loc['two_weekly'] = mse_tw[(mse_tw >= 1.96)].count()/mse_tw.shape[0] * 100
mse_neg_pct.loc['two_weekly'] = mse_tw[(mse_tw <= -1.96)].count()/mse_tw.shape[0] * 100
mse_pos_pct.loc['monthly'] = mse_m[(mse_m >= 1.96)].count()/mse_m.shape[0] * 100
mse_neg_pct.loc['monthly'] = mse_m[(mse_m <= -1.96)].count()/mse_m.shape[0] * 100
```
