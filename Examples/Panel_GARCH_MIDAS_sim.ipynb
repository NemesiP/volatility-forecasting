```python
import sys

sys.path.insert(1, 'C:/Users/peter/Desktop/volatility-forecasting/midas')

from volatility import Panel_GARCH_MIDAS
import pandas as pd
import numpy as np
import time
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
pm100 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/panel_garch_midas_100.csv')
pm200 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/panel_garch_midas_200.csv')
gm100 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/pmidas_sim100.csv')
gm200 = pd.read_csv('C:/Users/peter/Desktop/volatility-forecasting/results/pmidas_sim200.csv')
```

```python
alpha_100 = sm.nonparametric.KDEUnivariate(pm100.iloc[:, 4].values)
alpha_100.fit()
alpha_200 = sm.nonparametric.KDEUnivariate(pm200.iloc[:, 4].values)
alpha_200.fit()
alpha_100s = sm.nonparametric.KDEUnivariate(gm100.iloc[:, 4].values)
alpha_100s.fit()
alpha_200s = sm.nonparametric.KDEUnivariate(gm200.iloc[:, 4].values)
alpha_200s.fit()

beta_100 = sm.nonparametric.KDEUnivariate(pm100.iloc[:, 5].values)
beta_100.fit()
beta_200 = sm.nonparametric.KDEUnivariate(pm200.iloc[:, 5].values)
beta_200.fit()
beta_100s = sm.nonparametric.KDEUnivariate(gm100.iloc[:, 5].values)
beta_100s.fit()
beta_200s = sm.nonparametric.KDEUnivariate(gm200.iloc[:, 5].values)
beta_200s.fit()

beta0_100 = sm.nonparametric.KDEUnivariate(pm100.iloc[:, 1].values)
beta0_100.fit()
beta0_200 = sm.nonparametric.KDEUnivariate(pm200.iloc[:, 1].values)
beta0_200.fit()
beta0_100s = sm.nonparametric.KDEUnivariate(gm100.iloc[:, 1].values)
beta0_100s.fit()
beta0_200s = sm.nonparametric.KDEUnivariate(gm200.iloc[:, 1].values)
beta0_200s.fit()

beta1_100 = sm.nonparametric.KDEUnivariate(pm100.iloc[:, 2].values)
beta1_100.fit()
beta1_200 = sm.nonparametric.KDEUnivariate(pm200.iloc[:, 2].values)
beta1_200.fit()
beta1_100s = sm.nonparametric.KDEUnivariate(gm100.iloc[:, 2].values)
beta1_100s.fit()
beta1_200s = sm.nonparametric.KDEUnivariate(gm200.iloc[:, 2].values)
beta1_200s.fit()

theta_100 = sm.nonparametric.KDEUnivariate(pm100.iloc[:, 3].values)
theta_100.fit()
theta_200 = sm.nonparametric.KDEUnivariate(pm200.iloc[:, 3].values)
theta_200.fit()
theta_100s = sm.nonparametric.KDEUnivariate(gm100.iloc[:, 3].values)
theta_100s.fit()
theta_200s = sm.nonparametric.KDEUnivariate(gm200.iloc[:, 3].values)
theta_200s.fit()

fig , ax = plt.subplots(5, 1, figsize=(15, 15), tight_layout=True)

ax[0].plot(alpha_100.support, alpha_100.density, lw = 3, label = 'T = 100 & N = 100', zorder = 10)
ax[0].plot(alpha_200.support, alpha_200.density, lw = 3, label = 'T = 200 & N = 100',marker = '^', markevery = 10, zorder = 10)
#ax[0].plot(alpha_100s.support, alpha_100s.density, lw = 3, label = 'T = 100 & N = 1', zorder = 10)
#ax[0].plot(alpha_200s.support, alpha_200s.density, lw = 3, label = 'T = 200 & N = 1', zorder = 10)
ax[0].set_title(r'$\alpha$'+" (Act = 0.06) parameter's density from different samples sizes")
#ax[0].set_xlim((np.min(gm100.iloc[:, 4].values), 0.2))
ax[0].legend(loc = 'best')
ax[0].grid(True, zorder = -5)

ax[1].plot(beta_100.support, beta_100.density, lw = 3, label = 'T = 100 & N = 100', zorder = 10)
ax[1].plot(beta_200.support, beta_200.density, lw = 3, label = 'T = 200 & N = 100',marker = '^', markevery = 10, zorder = 10)
#ax[1].plot(beta_100s.support, beta_100s.density, lw = 3, label = 'T = 100 & N = 1', zorder = 10)
#ax[1].plot(beta_200s.support, beta_200s.density, lw = 3, label = 'T = 200 & N = 1', zorder = 10)
ax[1].set_title(r'$\beta$'+" (Act = 0.8) parameter's density from different samples sizes")
#ax[1].set_xlim((0.4, np.max(gm100.iloc[:, 5].values)))
ax[1].legend(loc = 'best')
ax[1].grid(True, zorder = -5)

ax[2].plot(beta0_100.support, beta0_100.density, lw = 3, label = 'T = 100 & N = 100', zorder = 10)
ax[2].plot(beta0_200.support, beta0_200.density, lw = 3, label = 'T = 200 & N = 100',marker = '^', markevery = 10, zorder = 10)
#ax[2].plot(beta0_100s.support, beta0_100s.density, lw = 3, label = 'T = 100 & N = 1', zorder = 10)
#ax[2].plot(beta0_200s.support, beta0_200s.density, lw = 3, label = 'T = 200 & N = 1', zorder = 10)
ax[2].set_title(r'$\beta_0$'+" (Act = 0.1) parameter's density from different samples sizes")
#ax[2].set_xlim((np.min(gm100.iloc[:, 1].values), np.max(gm100.iloc[:, 1].values)))
ax[2].legend(loc = 'best')
ax[2].grid(True, zorder = -5)

ax[3].plot(beta1_100.support, beta1_100.density, lw = 3, label = 'T = 100 & N = 100', zorder = 10)
ax[3].plot(beta1_200.support, beta1_200.density, lw = 3, label = 'T = 200 & N = 100', marker = '^', markevery = 10, zorder = 10)
#ax[3].plot(beta1_100s.support, beta1_100s.density, lw = 3, label = 'T = 100 & N = 1', zorder = 10)
#ax[3].plot(beta1_200s.support, beta1_200s.density, lw = 3, label = 'T = 200 & N = 1', zorder = 10)
ax[3].set_title(r'$\beta_1$'+" (Act = 0.3) parameter's density from different samples sizes")
#ax[3].set_xlim((np.min(gm100.iloc[:, 2].values), np.max(gm100.iloc[:, 2].values)))
ax[3].legend(loc = 'best')
ax[3].grid(True, zorder = -5)

ax[4].plot(theta_100.support, theta_100.density, lw = 3, label = 'T = 100 & N = 100', zorder = 10)
ax[4].plot(theta_200.support, theta_200.density, lw = 3, label = 'T = 200 & N = 100',marker = '^', markevery = 10, zorder = 10)
#ax[4].plot(theta_100s.support, theta_100s.density, lw = 3, label = 'T = 100 & N = 1', zorder = 10)
#ax[4].plot(theta_200s.support, theta_200s.density, lw = 3, label = 'T = 200 & N = 1', zorder = 10)
ax[4].set_title(r'$\theta$'+" (Act = 4.0) parameter's density from different samples sizes")
#ax[4].set_xlim((np.min(gm100.iloc[:, 3].values), np.max(gm100.iloc[:, 3].values)))
ax[4].legend(loc = 'best')
ax[4].grid(True, zorder = -5)

plt.savefig('C:/Users/peter/Desktop/volatility-forecasting/results/panel_garch_midas_sim.png')
plt.show()
```
